```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

```python
df = df.drop(['EmployeeNumber', 'Over18','EmployeeCount','StandardHours'],axis=1)
df.head()
```

```python
df.shape
```

```python
y = df['Attrition']
X = df.drop(['Attrition'],axis=1)
```

```python
X
```

```python
from sklearn import preprocessing
```

```python
lb = preprocessing.LabelBinarizer()
y = lb.fit_transform(y)
y
```

```python
X.info()
```

```python
X.describe()
```

```python
X.select_dtypes(['object']).head()
```

```python
# I should have used forloop/vectorizatin here
ind_BusinessTravel = pd.get_dummies(X['BusinessTravel'], prefix='BusinessTravel')
ind_Department = pd.get_dummies(X['Department'], prefix='Department')
ind_EducationField = pd.get_dummies(X['EducationField'], prefix='EducationField')
ind_Gender = pd.get_dummies(X['Gender'], prefix='Gender')
ind_JobRole = pd.get_dummies(X['JobRole'], prefix='JobRole')
ind_MaritalStatus = pd.get_dummies(X['MaritalStatus'], prefix='MaritalStatus')
ind_OverTime = pd.get_dummies(X['OverTime'], prefix='OverTime')
```

```python
X1 = pd.concat([ind_BusinessTravel, ind_Department, ind_EducationField, ind_Gender, 
                 ind_JobRole, ind_MaritalStatus, ind_OverTime])
```

```python
X1.head()
```

```python
X.select_dtypes(['int64']).head()
```

```python
X1 = pd.concat([ind_BusinessTravel, ind_Department, ind_EducationField, ind_Gender, 
                 ind_JobRole, ind_MaritalStatus, ind_OverTime, X.select_dtypes(['int64'])], axis=1)
```

```python
X1.head()
```

# Applying normal tree

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X1,y)
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(random_state=42)
```

```python
clf.fit(X_train,y_train)
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data) 
graph.render("hr") 
```

### Training Score

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(y_train,clf.predict(X_train))
```

```python
print(classification_report(y_train,clf.predict(X_train)))
```

```python
confusion_matrix(y_train, clf.predict(X_train))
```

### Testing Score

```python
accuracy_score(y_test,clf.predict(X_test))
```

```python
print(classification_report(y_test,clf.predict(X_test)))
```

```python
confusion_matrix(y_test, clf.predict(X_test))
```
